```python
#1D random walk

import math 
import matplotlib.pyplot as plt 
n = 0
n_plot = []
time_plot = []
time = 0
while n < 86:
    n_plot.append(n)
    time_plot.append(time)
    time = time + 1/(2**(2*n)) * 2*n * math.factorial(2*n) / (math.factorial(n))**2 
    n = n + 1
print(time)
plt.plot(n_plot,time_plot)
plt.show()

```

```python
#2D random walk

import math 
import matplotlib.pyplot as plt 
n = 0
n_plot = []
time_plot = []
time = 0
while n < 86:
    m = 0
    n_plot.append(n)
    time_plot.append(time)

    while m < n+1:
    
        time = time + 1/(4**(2*n)) * 2*n * math.factorial(2*n) / (math.factorial(m)*math.factorial(n-m))**2
        m = m+ 1
    n = n + 1
print(time)
plt.plot(n_plot,time_plot)
plt.show()
```

```python
#3D random walk

import math 
import matplotlib.pyplot as plt 
n = 0
n_plot = []
time_plot = []
time = 0
while n < 86:
    m = 0
    n_plot.append(n)
    time_plot.append(time)

    while m < n+1:
        k = 0
        while k < n-m + 1:
            
            time = time + 1/(6**(2*n)) * 2*n * math.factorial(2*n) / (math.factorial(k)*math.factorial(m)*math.factorial(n-m-k))**2
            k = k + 1
        m = m+ 1
    n = n + 1
print(time)
plt.plot(n_plot,time_plot)
plt.show()
```

```python
#4D random walk

import math 
import matplotlib.pyplot as plt 
n = 0
n_plot = []
time_plot = []
time = 0
while n < 86:
    m = 0
    n_plot.append(n)
    time_plot.append(time)

    while m < n+1:
        k = 0
        while k < n-m + 1:
            l = 0
            while l < n-m-k+1:
                time = time + 1/(8**(2*n)) * 2*n * math.factorial(2*n) / (math.factorial(l)*math.factorial(k)*math.factorial(m)*math.factorial(n-m-k-l))**2
                l = l + 1
            k = k + 1
        m = m+ 1
    n = n + 1
print(time)
plt.plot(n_plot,time_plot)
plt.show()

```
